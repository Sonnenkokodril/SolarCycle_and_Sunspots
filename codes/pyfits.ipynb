# Use of PyFits

```python
#Importing libraries
import numpy as np
%pylab inline
import matplotlib.pylab as plt
#PyFITS library
import pyfits
```

```python
#==================================================================
#Parameters
#==================================================================
#Fold of data
datafold = "../data/"
```

```python
#Loading data
filename = "%s/hmi.ic_45s.2014.10.28_00_01_30_TAI.continuum.fits"%(datafold)
sun, header = pyfits.getdata( filename, 0, header=True )
```

```python
#Plotting
plt.figure( figsize = (8,8) )
plt.imshow( sun, cmap="hot" )
plt.show()
```

```python
#Separating region with spots
sun_spot = sun[2300:2800,300:800]
plt.figure( figsize = (8,8) )
plt.imshow( sun_spot, cmap="hot" )
```
